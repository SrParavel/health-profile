```python
import pandas as pd

df = pd.read_csv("heart_2022.csv")
df.head(5)
```

```python
df.info()
```

```python
df = df.drop(columns="State")
df = df.drop(columns="RaceEthnicityCategory")

df["PhysicalHealthDays"] = df["PhysicalHealthDays"].astype(int)
df["MentalHealthDays"] = df["MentalHealthDays"].astype(int)
df["SleepHours"] = df["SleepHours"].astype(int)
df.info()
```

```python
get_uniques(df, df.columns)
```

# Encoding

```python
# Seleccionar solo las columnas categÃ³ricas (tipo objeto)
cat_cols = df.select_dtypes(include="object").columns

# Detectar columnas que solo tienen "Yes" y "No"
yes_no_cols = []
for col in cat_cols:
    valores_unicos = df[col].dropna().unique()
    if set(valores_unicos) <= {"Yes", "No"}:
        yes_no_cols.append(col)

# Convertir "Yes" a 1 y "No" a 0
for col in yes_no_cols:
    df[col] = (df[col] == "Yes").astype(int)
```

```python
df.info()
```

```python
# Seleccionamos las columnas categÃ³ricas restantes

cat_cols = df.select_dtypes(include="object").columns

#Specific encoding
general_health_map = {
    'Excellent': 4,
    'Very good': 3,
    'Good': 2,
    'Fair': 1,
    'Poor': 0
}

last_checkup_map = {
    'Within past year (anytime less than 12 months ago)': 3,
    'Within past 2 years (1 year but less than 2 years ago)': 2,
    'Within past 5 years (2 years but less than 5 years ago)': 1,
    '5 or more years ago': 0
}

removed_teeth_map = {
    'None of them': 0,
    '1 to 5': 1,
    '6 or more, but not all': 2,
    'All': 3
}

had_diabetes_map = {
    'No': 0,
    'No, pre-diabetes or borderline diabetes': 1,
    'Yes, but only during pregnancy (female)': 2,
    'Yes': 3
}

smoker_status_map = {
    'Never smoked': 0,
    'Former smoker': 1,
    'Current smoker - now smokes some days': 2,
    'Current smoker - now smokes every day': 3
}

ecig_usage_map = {
    'Never used e-cigarettes in my entire life': 0,
    'Not at all (right now)': 1,
    'Use them some days': 2,
    'Use them every day': 3
}

age_map = {
    'Age 18 to 24': 21,
    'Age 25 to 29': 27,
    'Age 30 to 34': 32,
    'Age 35 to 39': 37,
    'Age 40 to 44': 42,
    'Age 45 to 49': 47,
    'Age 50 to 54': 52,
    'Age 55 to 59': 57,
    'Age 60 to 64': 62,
    'Age 65 to 69': 67,
    'Age 70 to 74': 72,
    'Age 75 to 79': 77,
    'Age 80 or older': 85
}

tetanus_map = {
    'Yes, received Tdap': 3,
    'Yes, received tetanus shot, but not Tdap': 2,
    'Yes, received tetanus shot but not sure what type': 1,
    'No, did not receive any tetanus shot in the past 10 years': 0
}

covid_map = {
    'No': 0,
    'Tested positive using home test without a health professional': 1,
    'Yes': 2
}

# Aplicar los mapeos al DataFrame
df["Sex"] = (df["Sex"] == "Male").astype(int)
df['GeneralHealth'] = df['GeneralHealth'].map(general_health_map).astype(int)
df['LastCheckupTime'] = df['LastCheckupTime'].map(last_checkup_map).astype(int)
df['RemovedTeeth'] = df['RemovedTeeth'].map(removed_teeth_map).astype(int)
df['HadDiabetes'] = df['HadDiabetes'].map(had_diabetes_map).astype(int)
df['SmokerStatus'] = df['SmokerStatus'].map(smoker_status_map).astype(int)
df['ECigaretteUsage'] = df['ECigaretteUsage'].map(ecig_usage_map).astype(int)
df['AgeCategory'] = df['AgeCategory'].map(age_map).astype(int)
df['TetanusLast10Tdap'] = df['TetanusLast10Tdap'].map(tetanus_map).astype(int)
df['CovidPos'] = df['CovidPos'].map(covid_map).astype(int)


get_uniques(df, df.columns)
```
